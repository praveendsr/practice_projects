```python
import numpy as np
import pandas as pd
import scipy.stats as stats
ds=pd.read_csv('role_website.csv')
ds
```

```python
df=pd.DataFrame(ds)
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.dtypes
```

```python
#create contingency table

df_crosstab=pd.crosstab(df['role'],df['website'],margins=True)
print(df_crosstab)
```

```python
df_crosstab.columns=['facebook','linkedin','row_totals']
df_crosstab.index=['student','teacher','col_totals']
df_crosstab
```

```python
#get the observed table
observed=df_crosstab.iloc[0:2,0:2]
observed
```

```python
#expected table
#E=(row total*column total)/grand total
#E1=13*16/34
#E1
#E2=13*18/34

expected=np.outer(df_crosstab['row_totals'][0:2],df_crosstab.loc['col_totals'][0:2])/34
expected
```

```python
#calcuate the Chi square value and critical value

"""Chi square formula:
    chi square=total of [observed-expected)sq2]/expected
    Note: we call.sum() twice: once tot get the dolumn sums
    and a second time to add the column sums together,
    returning the sum of the entire 2D table."""
```

```python
chi_squared_stat=(((observed-expected)**2)/expected).sum().sum()
""" Note: we call.sum() twice: once to get the column sums and a returning the sum of the entire 2D table"""
print(chi_squared_stat)
```

```python
"""Find the critial value for onfidence of 95% and degree of freedom(df) of 1
why df=1?
degree of freedom formula:
    df=(total rows-1)*(total column-1)
    =(2-1)*(2-1)
    =1*1=1"""
```

```python
#above the very complicted way.
```

```python
# Methodology 2: calculate chi square using scipy.stat library

chi2,p,dof,expected=stats.chi2_contingency(observed=observed)

#you can check the retun data by calling it.
#the return data includes: chi_squared_stat, p_value, df expected_crosstab
#print('stresult=',stresult)

print('Chi square test=',chi2)
print("p_value=",p)
print("dof=",dof)
print("expected=", expected)

"""from this we interpret the p value is 0.16 that is less tha  0.05->>significance value. Hence null hypothesis is rejected and shows the two variables are dependent on each other."""
```

```python
#singficance value
alpha=0.05
#determine whether to reject or keep your null hypothesis
print('singinficance level alpha=%.2f, p=%.2f'%(alpha,p))
if p >= alpha:
    print('variables are not associated and independent(accept H0-Null hypothesis)')
else:
    print('variables are associated or dependent(reject H0 & accept H1')
```
